```python
#   We need to create a solution for the quadratic equation 
#       ax2 + bx + c = 0 (a, b, c ∈ R)
import cmath

def q_solve(a, b, c):
    D = b**2 - 4*a*c  # discriminant (can now be complex too)
    
    x1 = (-b + cmath.sqrt(D)) / (2*a)
    x2 = (-b - cmath.sqrt(D)) / (2*a)
    
    return x1, x2

# Tests
print(q_solve(1, -5, 6))        # real coefficients:    (3.0, 2.0)
print(q_solve(1, 0, 1))         # negative D:           (1j, -1j)
print(q_solve(1+0j, 2+1j, 1))   # complex coefficients: complex results
```

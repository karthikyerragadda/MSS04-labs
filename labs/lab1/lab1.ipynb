```python
#   We need to create a solution for the quadratic equation 
#       ax2 + bx + c = 0 (a, b, c ∈ R)
import math


def q_solve(a, b, c):
    D = b**2 - 4*a*c  # discriminant
    
    if D > 0:
        x1 = (-b + math.sqrt(D)) / (2*a)
        x2 = (-b - math.sqrt(D)) / (2*a)
        return x1, x2
    elif D == 0:
        x = -b / (2*a)
        return x
    else:
        return "no real solution"

print(q_solve(1, -5, 6))   # x²-5x+6=0 → should give 3.0, 2.0
print(q_solve(1, -2, 1))   # x²-2x+1=0 → should give 1.0
print(q_solve(1, 0, 1))    # x²+1=0    → no real solution
```

```python
# Exercise 2
# Heaviside function
def H(z):
    if z >= 0:
        return 1
    else:
        return 0

# Perceptron output
def output(x1, x2, w1, w2, b):
    z = w1*x1 + w2*x2 + b
    return H(z)

# Perceptron learning algorithm
def p_learn(X, Y, w1, w2, b, eta):
    converged = False
    while not converged:
        converged = True
        for (x1, x2), y in zip(X, Y):
            z = w1*x1 + w2*x2 + b
            if y == 1 and z < 0:       # should be A+, but wrong
                w1 += eta * x1
                w2 += eta * x2
                b  += eta
                converged = False
            elif y == 0 and z >= 0:    # should be A-, but wrong
                w1 -= eta * x1
                w2 -= eta * x2
                b  -= eta
                converged = False
    return w1, w2, b

# OR dataset
X = [[1,1],[1,0],[0,1],[0,0]]
Y_OR = [1, 1, 1, 0]

# Train
w1, w2, b = p_learn(X, Y_OR, 0, 0, 0, 1)
print(f"Learned weights: w1={w1}, w2={w2}, b={b}")

# Verify all outputs
for (x1, x2), y in zip(X, Y_OR):
    print(f"output({x1},{x2}) = {output(x1,x2,w1,w2,b)}, expected {y}")
```

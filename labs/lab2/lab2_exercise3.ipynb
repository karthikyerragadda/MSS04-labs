```python
## a) Let the logical term (X ∧¬Y)∨(¬(Z ∨Y)∨X) be given.
#    (i) Define a Python function l_term(X,Y,Z) which computes the truth value
 #   of the logical term using logical operators and, or, not. Use the inputs
  #  (True,True,False), (False,True,False), (False,False,True) for
   # (X,Y,Z).
```

```python
def l_term(X, Y, Z):
    return (X and not Y) or (not (Z or Y) or X)

# Test the three inputs
print(l_term(True,  True,  False))  # → True
print(l_term(False, True,  False))  # → False
print(l_term(False, False, True))   # → False
```

```python
# Exercise-3-b)

from sympy import symbols
from sympy.logic.boolalg import (Implies, Equivalent, Xor,
                                  truth_table, simplify_logic)

# ── i) ──────────────────────────────────────────────────────────────
A, B = symbols('A B')
expr_i = (A | ~B) & ~(~A & B)

# ── ii) ─────────────────────────────────────────────────────────────
x1, x2 = symbols('x1 x2')
expr_ii = Implies(~x1, x2) | (x1 & ~x2)

# ── iii) ────────────────────────────────────────────────────────────
X, Y, Z = symbols('X Y Z')
expr_iii = Xor(X & ~Y & Z,  Xor(~X & Y,  Y & ~Z))

# ── iv) ─────────────────────────────────────────────────────────────
u, v, w = symbols('u v w')
expr_iv = Implies(Equivalent(u & v, v & w),  Xor(u, ~w))

# ── v) ──────────────────────────────────────────────────────────────
a, b, c, d = symbols('a b c d')
expr_v = Implies(Xor(~(a | ~b), ~c | d),  ~a & b & ~c)

# ── vi) ─────────────────────────────────────────────────────────────
X1, X2, X3, X4 = symbols('X1 X2 X3 X4')
expr_vi = Equivalent(X1 & X2,  ~X3 | X4)


# ── Helper: pretty-print truth table ────────────────────────────────
def print_tt(expr, variables):
    print("Input(s) | Value")
    print("-" * 20)
    for vals, result in truth_table(expr, variables):
        print(f"  {list(vals)}  →  {result}")

# ── Helper: simplify and show result ────────────────────────────────
def show_simplified(label, expr):
    simplified = simplify_logic(expr)
    print(f"\n{label}  simplified:  {simplified}")
    return simplified


# ── Run everything ───────────────────────────────────────────────────
cases = [
    ("i",   expr_i,   [A, B]),
    ("ii",  expr_ii,  [x1, x2]),
    ("iii", expr_iii, [X, Y, Z]),
    ("iv",  expr_iv,  [u, v, w]),
    ("v",   expr_v,   [a, b, c, d]),
    ("vi",  expr_vi,  [X1, X2, X3, X4]),
]

for label, expr, variables in cases:
    print(f"\n{'='*40}")
    print(f"Term {label} — original truth table:")
    print_tt(expr, variables)
    simplified = show_simplified(label, expr)
    print(f"\nTerm {label} — simplified truth table:")
    print_tt(simplified, variables)
```

```python
# Exercise 3-c)

from sympy.logic.inference import satisfiable

# ── Check satisfiability + collect all models ────────────────────────
def check_sat(label, expr, variables):
    print(f"\n{'='*40}")
    print(f"Term {label}:  {expr}")

    # Basic check — one model or False
    result = satisfiable(expr)
    if result == False:
        print("  → UNSATISFIABLE (contradiction)")
        return

    print(f"  → SATISFIABLE, one model: {result}")

    # All models
    all_m = list(satisfiable(expr, all_models=True))
    
    # satisfiable() returns {False: False} for tautologies when all_models=True
    # In that case every valuation is a model, so we handle it separately
    if all_m == [{False: False}]:
        n = 2 ** len(variables)
        print(f"  → TAUTOLOGY — all {n} valuations are models")
        return

    print(f"  → All {len(all_m)} model(s):")
    for m in all_m:
        print(f"     {m}")


# Re-use expressions and variables from b)
cases = [
    ("i",   expr_i,   [A, B]),
    ("ii",  expr_ii,  [x1, x2]),
    ("iii", expr_iii, [X, Y, Z]),
    ("iv",  expr_iv,  [u, v, w]),
    ("v",   expr_v,   [a, b, c, d]),
    ("vi",  expr_vi,  [X1, X2, X3, X4]),
]

for label, expr, variables in cases:
    check_sat(label, expr, variables)
```

```python
from sympy import symbols
from sympy.logic.boolalg import Implies, Equivalent, Xor
from sympy.logic.inference import satisfiable

# ── Redefine all symbols and expressions ────────────────────────────
A, B         = symbols('A B')
x1, x2       = symbols('x1 x2')
X, Y, Z      = symbols('X Y Z')
u, v, w      = symbols('u v w')
a, b, c, d   = symbols('a b c d')
X1,X2,X3,X4  = symbols('X1 X2 X3 X4')

expr_i   = (A | ~B) & ~(~A & B)
expr_ii  = Implies(~x1, x2) | (x1 & ~x2)
expr_iii = Xor(X & ~Y & Z, Xor(~X & Y, Y & ~Z))
expr_iv  = Implies(Equivalent(u & v, v & w), Xor(u, ~w))
expr_v   = Implies(Xor(~(a | ~b), ~c | d), ~a & b & ~c)
expr_vi  = Equivalent(X1 & X2, ~X3 | X4)

cases = [
    ("i",   expr_i,   [A, B]),
    ("ii",  expr_ii,  [x1, x2]),
    ("iii", expr_iii, [X, Y, Z]),
    ("iv",  expr_iv,  [u, v, w]),
    ("v",   expr_v,   [a, b, c, d]),
    ("vi",  expr_vi,  [X1, X2, X3, X4]),
]

# ── Check satisfiability + collect all models ────────────────────────
def check_sat(label, expr, variables):
    print(f"\n{'='*40}")
    print(f"Term {label}:  {expr}")

    result = satisfiable(expr)
    if result == False:
        print("  → UNSATISFIABLE (contradiction)")
        return

    print(f"  → SATISFIABLE, one model: {result}")

    all_m = list(satisfiable(expr, all_models=True))

    if all_m == [{False: False}]:
        n = 2 ** len(variables)
        print(f"  → TAUTOLOGY — all {n} valuations are models")
        return

    print(f"  → All {len(all_m)} model(s):")
    for m in all_m:
        print(f"     {m}")

# ── Run ──────────────────────────────────────────────────────────────
for label, expr, variables in cases:
    check_sat(label, expr, variables)
```

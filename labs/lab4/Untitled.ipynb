```python
def output(self, x, y):
    a = x                          # start with the input
    for layer in self.lay_list:    # go through each layer
        a = layer.activations(a)   # output of this layer becomes input of next
    return a                       # final output is y_hat
```

```python
def C_prime(self, y_hat, y):
    return y_hat - y
```

```python
def nabla_C(self, x, y):
    # --- FORWARD PASS: save z and a at every layer ---
    zs = []       # list of z values per layer
    activations = [x]   # list of activations, starting with input x

    a = x
    for layer in self.lay_list:
        z = np.dot(layer.weights, a) + layer.biases
        zs.append(z)
        a = sigmoid(z)
        activations.append(a)

    # --- BACKWARD PASS ---
    # Empty lists to store gradients (one entry per layer)
    nabla_b = [np.zeros(l.biases.shape) for l in self.lay_list]
    nabla_W = [np.zeros(l.weights.shape) for l in self.lay_list]

    # sigmoid derivative
    def sigmoid_prime(z):
        return sigmoid(z) * (1 - sigmoid(z))

    # Start at output layer (index -1)
    y_hat = activations[-1]
    delta = self.C_prime(y_hat, y) * sigmoid_prime(zs[-1])
    nabla_b[-1] = delta
    nabla_W[-1] = np.outer(delta, activations[-2])

    # Work backwards through remaining layers
    for l in range(2, self.n_layer):
        z = zs[-l]
        delta = np.dot(self.lay_list[-l+1].weights.T, delta) * sigmoid_prime(z)
        nabla_b[-l] = delta
        nabla_W[-l] = np.outer(delta, activations[-l-1])

    return nabla_b, nabla_W
```

```python
def learn(self, training_data, eta=0.1, epochs=10):
    for epoch in range(epochs):
        for x, y in training_data:
            nabla_b, nabla_W = self.nabla_C(x, y)
            for i, layer in enumerate(self.lay_list):
                layer.weights -= eta * nabla_W[i]
                layer.biases  -= eta * nabla_b[i]
```

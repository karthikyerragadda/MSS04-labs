```python
##Exercise 11. (Gradient descent)
#1) Define a Python function grad_descent() with the parameters x_start, f,
# f_prime, and the optional parameters eta (default value 0.1), iter (default
# value 10), where x_start is the starting value of the recursion, f is a function
# of one or more variables that is to be minimized, f_prime its derivative, re
# spectively gradient, eta the constant in the recurrence equation and iter the
# total number of iterations. In each step of iteration print out the step number j
#  (j = 1,...,iter), the values of x(j-1), f_prime(x(j-1)), x(j) and f(x(j))
# (use a numpy array for x(j) of appropriate size).
```

```python
#11-1
import numpy as np
def grad_descent(x_start,f,f_prime,eta=0.1,iter=10):
    x=x_start
    for j in range(iter):
        x_star= x - eta * f_prime(x) # iteration step
        print(f"{j+1} {np.round(x,2)}          {np.round(f_prime(x),2)}          {np.round(x_star,2)}       {np.round(f(x_star),2)}")
        x=x_star
        
```

```python
#11-2-a
def f(x):       return (x - 2)**2 + 1
def f_prime(x):      return 2 * (x - 2)
for eta in [0.1, 0.5, 1.0, 1.1]:
    print(f"\n  ── eta = {eta}, x_start = 5.0 ──")
    grad_descent(5.0, f, f_prime, eta=eta, iter=10)
 
```

```python
#11-2-b
def f(x):
    return 0.5*(x - 2)*(x-2)*x*(x + 2)
 
def f_prime(x):
    return 2*x*x*x - 3*x*x - 4*x  + 4
for x in [0.0, 3.0]:
    for eta in [0.1, 0.2, 1.0]:
        print(f"\n  ── eta = {eta}, x_start = {x} ──")
        grad_descent(x, f, f_prime, eta=eta, iter=10)
```

```python
#11-2-c
def f(x):
    return (x[0]-2)**2 + (x[1]-2)**2 + 1
def f_prime(x):
    return np.array([2*x[0]-1,2*x[1]-4])
grad_descent(np.array([3,2]),f,f_prime, iter=20)
```

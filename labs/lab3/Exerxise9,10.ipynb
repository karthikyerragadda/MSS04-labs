```python
import matplotlib.pyplot as plt
import numpy as np


def sigmoid(x):
    return(1/(1+np.exp(-x)))
x=np.linspace(-2,2,101)
y=sigmoid(x)

plt.figure(figsize=(8,5))
plt.subplot(321)

plt.plot(x,y,label="sigmoid")
plt.title("Sigmoid")

plt.grid()
plt.xlabel('X')
plt.ylabel('Y')

#plt.legend()
plt.show()
np.tanh(2)


# Tanh function
y=np.tanh(x)

plt.figure(figsize=(8,5))
plt.subplot(322)

plt.plot(x,y,label="Tanh")
plt.title("Tanh")

plt.grid()
plt.xlabel('X')
plt.ylabel('Y')

#plt.legend()
plt.show()

```
